```python
two_word = dict()
general_dict = []

for i, data in enumerate(data_generation):
    for k,v in data.items():
        split_k = k.split(" ")
        if len(split_k)>1:
            join_k = '_'.join(split_k)
            two_word[k] = join_k
            general_dict.append({join_k : v})
        else:
            general_dict.append({k : v})

            
```

```python
data_df['input'].replace(to_replace=two_word,inplace=True)
test_df['input'].replace(to_replace=two_word,inplace=True)
```

```python
word_dict = dict()

for i, data in enumerate(general_dict):
    for k,v in data.items():
        if k not in word_dict:
            word_dict[k] = []
        if isinstance(v, list):
            for value in v:
                word_dict[k].append(value)
        else:
            word_dict[k].append(v)
```

```python
from tqdm import tqdm
import random
generated_sentences_1 = []
generated_sentences_2 = []
generated_sentences_3 = []
generated_sentences_4 = []
generated_sentences_5 = []
target_sentences = []

for i, row in tqdm(data_df.iterrows()):
    sentence = row['target']
    split_sentence = sentence.split(" ")
    gen_sentence_1 = []
    gen_sentence_2 = []
    gen_sentence_3 = []
    gen_sentence_4 = []
    gen_sentence_5 = []

    count = 0
    for word in split_sentence:
        choose = random.choice([True, False])
        if (word in word_dict) and choose and (count < 2):
            if count < 2:
                gen_sentence_1.append(random.choice(word_dict[word]))
                gen_sentence_2.append(random.choice(word_dict[word]))
                gen_sentence_3.append(random.choice(word_dict[word]))
                gen_sentence_4.append(random.choice(word_dict[word]))
                gen_sentence_5.append(random.choice(word_dict[word]))
                count += 1
        else:
            gen_sentence_1.append(word)
            gen_sentence_2.append(word)
            gen_sentence_3.append(word)
            gen_sentence_4.append(word)
            gen_sentence_5.append(word)
    
    gen_sentence_1 = ' '.join(gen_sentence_1)
    gen_sentence_2 = ' '.join(gen_sentence_2)
    gen_sentence_3 = ' '.join(gen_sentence_3)
    gen_sentence_4 = ' '.join(gen_sentence_4)
    gen_sentence_5 = ' '.join(gen_sentence_5)
    
    generated_sentences_1.append(gen_sentence_1)
    generated_sentences_2.append(gen_sentence_2)
    generated_sentences_3.append(gen_sentence_3)
    generated_sentences_4.append(gen_sentence_4)
    generated_sentences_5.append(gen_sentence_5)
    
    target_sentences.append(sentence)
```
