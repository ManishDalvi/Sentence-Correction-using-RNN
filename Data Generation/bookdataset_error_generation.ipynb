```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_df = pd.read_csv("all_books.csv")
```

```python
data_df, test_df = train_test_split(data_df, train_size=0.99)
```

```python
data_df = data_df.replace(to_replace=r'[^A-Za-z0-9 ]+', value='', regex=True)

test_df = test_df.replace(to_replace=r'[^A-Za-z0-9 ]+', value='', regex=True)
```

```python
from tqdm import tqdm
import random

generated_sentences_1 = []
generated_sentences_2 = []
generated_sentences_3 = []
generated_sentences_4 = []

target_sentences = []

CHARS = list('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ')
REMOVE_CHARS = '[#$%"\+@<=>!&,-.?:;()*\[\]^_`{|}~/\d\t\n\r\x0b\x0c]'

for i, row in tqdm(data_df.iterrows()):
    sentence = row['target']
    split_sentence = sentence.split(" ")
    gen_sentence_1 = []
    gen_sentence_2 = []
    gen_sentence_3 = []
    gen_sentence_4 = []
    
    count = 0
    for word in split_sentence:
        choose = random.choice([True, False])
        
        if choose and (count!= 3):
            # Delete letter
            error_word_list = []
            if len(word) > 1:
                letter_index = random.randrange(len(word))
                new_word = word[:letter_index] + word[letter_index+1:]
                error_word_list.append(new_word)
            else:
                error_word_list.append(word)
                
            # Substitute letter
            if len(word) > 1:
                letter_index = random.randrange(len(word))
                new_word = word[:letter_index] +  random.choice(CHARS) +word[letter_index+1:]
                error_word_list.append(new_word)
            else:
                error_word_list.append(word)
            
            # Add random letter
            if len(word) > 1:
                letter_index = random.randrange(len(word))
                new_word = word[:letter_index] + random.choice(CHARS) + word[letter_index:]
                error_word_list.append(new_word)
            else:
                error_word_list.append(word)
                
            # Switch two letters
            if len(word) > 1:
                letter_index = random.randrange(len(word)-1)
                new_word = word[:letter_index] + word[letter_index+1] + word[letter_index] + word[letter_index+2:]
                error_word_list.append(new_word)
            else:
                error_word_list.append(word)
                
            gen_sentence_1.append(random.choice(error_word_list))
            gen_sentence_2.append(random.choice(error_word_list))
            gen_sentence_3.append(random.choice(error_word_list))
            gen_sentence_4.append(random.choice(error_word_list))
                
            
            count += 1
            
        else:
            gen_sentence_1.append(word)
            gen_sentence_2.append(word)
            gen_sentence_3.append(word)
            gen_sentence_4.append(word)
    
    gen_sentence_1 = ' '.join(gen_sentence_1)
    gen_sentence_2 = ' '.join(gen_sentence_2)
    gen_sentence_3 = ' '.join(gen_sentence_3)
    gen_sentence_4 = ' '.join(gen_sentence_4)
    
    generated_sentences_1.append(gen_sentence_1)
    generated_sentences_2.append(gen_sentence_2)
    generated_sentences_3.append(gen_sentence_3)
    generated_sentences_4.append(gen_sentence_4)
    
    target_sentences.append(sentence)
```

```python
from tqdm import tqdm
import random

generated_sentences_1_test = []

target_sentences_test = []

CHARS = list('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ')
REMOVE_CHARS = '[#$%"\+@<=>!&,-.?:;()*\[\]^_`{|}~/\d\t\n\r\x0b\x0c]'

for i, row in tqdm(test_df.iterrows()):
    sentence = row['target']
    split_sentence = sentence.split(" ")
    gen_sentence_1 = []
    
    count = 0
    for word in split_sentence:
        choose = random.choice([True, False])
        value = random.choice([0,1,2,3])
        if choose and (count!= 3):
            
            if value == 0:
                # Delete letter
                if len(word) > 1:
                    letter_index = random.randrange(len(word))
                    new_word = word[:letter_index] + word[letter_index+1:]
                    gen_sentence_1.append(new_word)
                else:
                    gen_sentence_1.append(word)
            if value == 1:
                # Substitute letter
                if len(word) > 1:
                    letter_index = random.randrange(len(word))
                    new_word = word[:letter_index] +  random.choice(CHARS) +word[letter_index+1:]
                    gen_sentence_1.append(new_word)
                else:
                    gen_sentence_1.append(word)
            
            if value == 2:
                # Add random letter
                if len(word) > 1:
                    letter_index = random.randrange(len(word))
                    new_word = word[:letter_index] + random.choice(CHARS) + word[letter_index:]
                    gen_sentence_1.append(new_word)
                else:
                    gen_sentence_1.append(word)
            
            if value == 3:
                # Switch two letters
                if len(word) > 1:
                    letter_index = random.randrange(len(word)-1)
                    new_word = word[:letter_index] + word[letter_index+1] + word[letter_index] + word[letter_index+2:]
                    gen_sentence_1.append(new_word)
                else:
                    gen_sentence_1.append(word)
            count += 1
        else:
            gen_sentence_1.append(word)
    
    gen_sentence_1 = ' '.join(gen_sentence_1)
    
    generated_sentences_1_test.append(gen_sentence_1)
    
    target_sentences_test.append(sentence)
```

```python
df_1_novel = pd.DataFrame({
    'input':generated_sentences_1,
    'target':target_sentences
})

df_2_novel = pd.DataFrame({
    'input':generated_sentences_2,
    'target':target_sentences
})

df_3_novel = pd.DataFrame({
    'input':generated_sentences_3,
    'target':target_sentences
})

df_4_novel = pd.DataFrame({
    'input':generated_sentences_4,
    'target':target_sentences
})
```

```python
test_novel = pd.DataFrame({
    'input':generated_sentences_1_test,
    'target':target_sentences_test
})
```

```python
train_novel = [df_1_novel, df_2_novel, df_3_novel, df_4_novel]

train_novel = pd.concat(train_novel)
```

```python
train_novel = train_novel.sample(frac=1)
print("Before dropping duplicates", train_novel.shape)
train_novel.drop_duplicates()
print("After dropping duplicates", train_novel.shape)
```
